```python
import cv2
from deepface import DeepFace
```

```python
from PIL import Image
import requests
#url = 'https://github.com/nngasu4course/Face_Recognition/raw/main/elon.jpg'

url ='https://www.sheknows.com/wp-content/uploads/2018/08/Happy_woman_ih9lnw.jpeg'
image = Image.open(requests.get(url, stream=True).raw)
```

```python
import numpy as np
img = np.asarray(image)
```

```python
result = DeepFace.analyze(img, actions=['emotion'])
```
